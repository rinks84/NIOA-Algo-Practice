```python
# Required Libraries
import numpy  as np
import math
import random
import os
import matplotlib.pyplot as plt

#Generating Initial Population S with Dimension d and Iterations N
#S = 5
#d = 2
#N = 10
# Input Swarm Strength

S = int(input("Enter Wolf Strength in Integers = "))  # Input Wolf Strength
d = int(input("Enter the No of dimensions Required = ")) # Input dimension size
N = int(input("Enter the No of Iterations Required = ")) # Input No of Iterations

print("Let's Start with the Entered values of Swarm Strength =", S,"\nAnd No of Iterations as =", N)

# Initialize Population of Wolves
X=[]
rows, cols= S, d
for i in range(rows):
    col = []
    for j in range(cols):
        col.append(round((random.uniform(-5,5)),4))
    X.append(col)

X = np.array(X)
X = X.reshape(S,d)
print("Randomly Generated Initial Population, X\n", X)

# Starting With Iterations
for n in range(N):
    print("\nIteration No :", n)
    X_upt = []
    for s in range(S):
        print("\nWolf No :", s)
        X_init = X[s]      # Initial Position of Prey
        #X_upt = []         # Updated Wolves Position Vector 
        # Generating random values A
        a = 2 - n/N
        A=[]
        rows, cols = S, d
        for j in range(rows):
            for k in range(cols):
                r = random.random()
                A.append(round((2*a*r - a),2))
            A.append
        A = np.array(A)
        A = A.reshape(S,d)
        print("Randomly Generated Values of A\n", A)

        # Generating random values C
        C=[]
        rows, cols = S, d
        for i in range(rows):
            for j in range(cols):
                C.append(round((2*random.random()),2))
        C = np.array(C)
        C = C.reshape(S,d)
        print("Randomly Generated Values of C\n", C)

        # Calculating Fitness Function F1
        F1=[]
        rows = S
        for i in range(rows):
            F1.append(round((X[i][0]**2 + X[i][1]**2),4))
            F1.append
        print("\nCorresponding Fitness Values = ", F1)

        F1.sort()
        print("\nSorted Fitness Values\n", F1)

        Fα = F1[0]; print("\nFα", Fα)
        Fβ = F1[1]; print("Fβ", Fβ)
        Fδ = F1[2]; print("Fδ", Fδ)

        for i in range(rows):
            sol = round((X[i][0]**2 + X[i][1]**2),4)
            if(sol == Fα):
                Xα = X[i]; print("Xα", Xα)
            if(sol == Fβ):
                Xβ = X[i]; print("Xβ", Xβ)
            if(sol == Fδ):
                Xδ = X[i]; print("Xδ", Xδ)
            F1.append  
    
        # Calculating D Values
        Da = []; Db = []; Dc = [];
            
        for j in range(d):
            Da.append(round(abs(C[0][j]*Xα[j] - X_init[j]),4)); #print("Da[0] = ", Da)
            Db.append(round(abs(C[1][j]*Xα[j] - X_init[j]),4)); #print("Db[0] = ", Db)
            Dc.append(round(abs(C[2][j]*Xα[j] - X_init[j]),4)); #print("Dc[0] = ", Dc)
            
        print("Dα ", Da)
        print("Dβ ", Db)
        print("Dδ ", Dc)
            
        # Calculating X1, X2 and X3 Values
        X1 = []; X2 = []; X3 = [];
            
        for j in range(d):
            X1.append(round(Xα[j]-A[0][j]* Da[j],4)); #print("X1 = ", X1[j])
            X2.append(round(Xβ[j]-A[1][j]* Da[j],4)); #print("X2 = ", X2[j])
            X3.append(round(Xδ[j]-A[2][j]* Dc[j],4)); #print("X3 = ", X3[j])
            
        print("X1 ", X1)
        print("X2 ", X2)
        print("X3 ", X3)
        
        X1 = np.array(X1)
        X2 = np.array(X2)
        X3 = np.array(X3)
            
        X_upt.append((X1 + X2 + X3)/3); print("\nX_upt", X_upt)
                
    s+=1
    
n+=1

# Plotting Final Result with Initial Wolves Position
fig, ax= plt.subplots()
for i in range(S):
    plt.scatter(X[i][0], X[i][1], c ='r')
    plt.scatter(X_upt[i][0], X_upt[i][1], c ='g')
plt.show()
```
